```python
import pandas as pd
import numpy as np
import os
print(os.listdir(os.getcwd()))

def near(y1, y2):
    if np.abs(y1 - y2) < 15:
        return True
    return False

def make_rows(data: pd.DataFrame):
    data = data.sort_values(['upper_left_y', 'upper_left_x']).reset_index(drop=1)

    data_post = list(enumerate(data.itertuples()))
    result_rows = []
    for pos, d in data_post:
        _, prev_post = data_post[pos-1]
        if near(prev_post.upper_left_y, d.upper_left_y):
            if len(result_rows) > 0:
                try:
                    result_rows[-1] += [d]
                except:
                    print('*')
                    print(result_rows)
                    print(result_rows[-1])
                    print([d])
                    raise Exception()
            else:
                result_rows = [[d]]
        else:
            result_rows += [[d]]
    
    for i, r in enumerate(result_rows):
        try:
            result_rows[i] = list(sorted(result_rows[i], key=lambda x: x.upper_left_x))
        except:
            print(result_rows[i])
            print(result_rows[i].upper_left_x)
            raise Exception()
    result_text = []
    for i, r in enumerate(result_rows):
        
        result_text.append(' '.join([str(t.text) for t in r ]))
            
    return result_text
```
